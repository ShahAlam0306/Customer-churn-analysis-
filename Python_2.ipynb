```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
        
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
plt.figure(figsize = (3,5))
ax = sns.countplot(x = 'Churn' , data = df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.show()
```

```python
#73.46% customer has churned out
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='gender', data=df , hue = 'Churn' )
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

```

```python
sns.countplot(x='SeniorCitizen', data=df , hue = 'Churn' )
plt.title("churn by senior citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot ( x = 'tenure' , data = df , bins = 50 ,hue = 'Churn')
plt.show()
```

```python
sns.countplot(x='Contract', data=df , hue = 'Churn' )
plt.title("count of customer by contract")
plt.xlabel("Contract")
plt.ylabel("Count")
plt.show()
```

```python
# List of your columns
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the figure with subplots — 3 rows x 3 columns
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
axes = axes.flatten()  # flatten 2D array to 1D to loop easily

# Create countplots for each column
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i] , hue = df["Churn"])
    axes[i].set_title(f'{col} Distribution')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
#based on above graphs it shows the services used by customer and the reason of them churning out 
```
